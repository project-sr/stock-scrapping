# 변수별 최근 데이터 업데이트 코드

```python

#원자재 

from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from splinter.driver.webdriver import BaseWebDriver, WebDriverElement
import pandas as pd 
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import time


url = 'https://finance.daum.net/domestic/exchange'
options = Options()
options.add_argument('--start-maximized')

browser = BaseWebDriver()
driver = Chrome("C:/Users/glam/Desktop/chromedriver_win32/chromedriver.exe",chrome_options=options)
driver.get(url)


num = re.compile('[^0-9]')
aa = re.compile(',')

result = driver.page_source
soup = BeautifulSoup(result,"html.parser",from_encoding='utf-8')


matarial = soup.findAll('div',{'class':'tableB mt'})
matarial_name = [str(i).split(">")[1].split("<")[0] for i in matarial[0].findAll('a',{'class','txt'})]

price = []
num = re.compile('[^0-9]')
aa = re.compile(',')
i=0
for k in matarial[0].findAll('span',{'class','num'}):
    if i % 3 == 0:
        kk = str(k).split(">")[1]
        kkk = kk.split("<")[0]
        kkkk = aa.sub('',kkk)
        price.append(float(kkkk))
    i = i + 1 
    
CD = str(matarial[1].find('a',{'class','txt'})).split(">")[1].split("<")[0]
CD_price = float(str(matarial[1].find('span',{'class','num'})).split(">")[1].split("<")[0])

matarial_name.append(CD)
price.append(CD_price)

pd.DataFrame(data={'원자재':matarial_name, '가격':price})

```

# 환율 업데이트 코드

```python
# https://finance.naver.com/marketindex/exchangeDetail.nhn?marketindexCd=FX_USDKRW
# https://finance.naver.com/marketindex/exchangeDetail.nhn?marketindexCd=FX_EURKRW
# https://finance.naver.com/marketindex/exchangeDetail.nhn?marketindexCd=FX_JPYKRW
# https://finance.naver.com/marketindex/exchangeDetail.nhn?marketindexCd=FX_CNYKRW

from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from splinter.driver.webdriver import BaseWebDriver, WebDriverElement
import pandas as pd 
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import time

url = "https://finance.naver.com/marketindex/"
url_get = requests.get(url)
html = url_get.text
soup = BeautifulSoup(html,"html.parser",from_encoding='utf-8')

a = soup.find("div",{'class','market1'})
aa = re.compile(',')

country = []
price = []

j=0
for k in a.findAll("span",{"class","blind"}):
    if j % 3 ==0:
        country.append(str(k).split(">")[1].split("<")[0])
    j = j+1
        
for q in a.findAll("span",{"class","value"}):
    price.append(aa.sub('',str(q).split(">")[1].split("<")[0]))


pd.DataFrame(data={'환율국가':country, '가격':price})
```
